# Hierarchical Clustering

It is also known as aglomatic clustering technique 

in this technique we initially find out the distance between each data point and each datapoint is considered as a cluster in itself 

Now at each iteration the similar cluster merge with another cluster until one cluster or k -cluster is formed 

we will have different datapoints 

each datapoints will be considered as cluster in itself 

then we will create a matrix which will have the distance of each datapoint with other datapoint 

then we will combine the nearesrt datapoint into one particular group 

once we have these nearest datapoint then again we will recombine to another group and so on

so it will keep on combining different groups ,till the required number of clusters are reached  

This method does not require the number of cluster to be specified at the begining because in the begning each dayapoint is considered as cluster 
so we will start the process and will go until we get one cluster left 

so all the points will slowly and gradually be combined into single cluster 

we can subdivide or stop at any point of time when we actually achieve k cluster 

the distance connectivity between observation is the measure 
so we will find out the distance between each and evry datapoints 

so to find out each and evrry datapoint distance , we will need to have them scaled 
we can not have age ranging from 0-100 and income in corores because that will have a very large diffrenece in scale 

when we will have large diffreence in the scale , then the distnace which will be between the ages , will be dimnished due to the distance which would be present in the income 

so what will happen because age will have smaller distance so all the age data , all the datapoint which are closer in age will be grouped together and there will be no impact of amount 

so in this we will apply bottom - up approach

so each observation  start in its own cluster and the similarity or the distance between each cluster 

and then we merge the most similar ones , each operation until there is only one cluster left

niw we will be computing the distance between each and every cluster after every iteration 

so we can see that it is computationaly expensive 

so there is high computation is required 

and because there is such such a high need of computaional required 

it is feaible to use aggomerated clustering or herarchical clutering only when we have small amount of data 

if we have large dataset it will take lot of time to comute this

# Imprt libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.preprocessing import scale  # to scale the data 
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
iris = pd.read_csv("iris.csv")
```

```python
iris
```

Here we don'r ahve to classify anything we are just implimenting clustering algorithm 

we don't need target value also we don't need id value 

```python
iris.drop(['Id','Species'], axis=1, inplace=True)
```

```python
iris.head()
```

Now let's create the cluster out of the two columns 

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

Now we will scale the data

```python
iris_std = pd.DataFrame(scale(iris), columns=list(iris.columns))
```

```python
iris_std
```

Noww we can see that data is scaled now

```python
plt.figure(figsize=(10,5))
plt.xlabel('sample index')
plt.ylabel('distance')
z = linkage(iris_std,method='ward')
dendrogram(z,leaf_rotation=90,p = 5,color_threshold=10,leaf_font_size=10,truncate_mode='level')
plt.tight_layout()
```

z will create linkage for us 

so we have given linkage and dataset and the methode which we want to create in linkage

and then from the dendogram we want to print the linkage which we have created 

we can see that dendograms contain size of the clusters 

```python
iris_std=iris_std[["SepalLengthCm","SepalWidthCm"]]
iris_std.describe()
```

here we are taking only two columns and will plot the cluster digram

so now we will create the clusters 

```python
for n_clusters in range(2,20):
    cluster_model = AgglomerativeClustering(n_clusters=n_clusters, affinity='euclidean',linkage='ward')
    cluster_labels = cluster_model.fit_predict(iris_std)
    silhouette_avg = silhouette_score(iris_std,cluster_labels,metric='euclidean')
    print("For n_clusters =", n_clusters, 
          "The average silhouette_score is:", silhouette_avg)
```

So here we have iterated the for loop 

where we are iterating on the top of clusters and ranging from 2-19

and we ahve created cluster model object 

silhouete score closest to 1 is best silhoute score 

so we will seelct the maximum sihate score 

```python
silhouette_avg
```

```python
hclus=AgglomerativeClustering(n_clusters=3, affinity='euclidean',linkage='ward')
labels_hclus=hclus.fit_predict(iris_std)
iris['cluster_hclus']=labels_hclus
sns.lmplot(fit_reg=False,x="SepalLengthCm",y="SepalWidthCm",data=iris,hue='cluster_hclus')
```
