# K-means clustering

1. K is the number of clusters which we will be creating 

So k means clustering has k numbers of cluters 

2. The mechanism randomly initializes k random centroid in feature space 

3. At the time of intialization the k points are not at actual centroid of data 

4. the data points are assigned to nearest k points

5. the centroids are moved so that they are at the center of the current designated cluster
