# Sampling technique

- so when we are taking samples we have to apply several sampling technque
 - there are two type sampling technique which are available
    - Probability Sampling
    - Non probability sampling 

### Probability Sampling

- It involves random selection allowing us to make statistical infences about the whole group
- sampling is done by random selection of elements from population 

### Non- Probability Sampling

- It involves non- random selection based on convenience or other criteria allowing us to easily collect initial data 


### Type of Probability sampling 

#### 1.Simple random sampling 

- In this we take random value from the entire set, we pickout random elemnts from the entire set

#### 2. Systematic sampling 
- in this type of sampling we systematically pick a particular sequence of values from the entire data 
- ex: let's say we have value from 1 to 20 so we are picking up all the even data or multiple of 3

#### 3.Stratified random sampling
- the stratified sampling has the values big doubt, a clear cluster has formed 
- different cluster have been formed from the data 
- ex: cluster of old and young men have been taking seperately, another cluster of woman have been taken seperately and the cluster of children have been taken seperately
- and from these clusters people are picked out seperately 

#### 4.Cluster sampling 

- In this we create clusters out of entire data 
- we create different clusters from the data and we just pick out one cluster from the entire data and we use them for samples

 **Variance**
- In probability theory and statistics variance is the average of the squared difference from the mean
- Informally , it measures how far a set of numbers are spread out from their average value.

- there is difference between those two formula
  - in case of pupulation we use n in the denominator
  - while in case of samples we use n-1 in the denominator

**Standard deviation**
- It is the squared root of the variance 

- When we are calculating sample variance or standard deviation we are using n-1 instead of n 
- this is because of Besels correction 

### Bessels correction 

- Bessels correction refres to the the n-1 found in several formulas , including the sample variance and sample standard deviation formulas . thuis correction is made to correct for the fact that thses sample statistics tend to underestimate the acual parameter found in the population 
-  The Standard deviation And variance which we are calculating for the sample is kind of not considering the entire picture and might be little lowe than the actual variance and standard deviation of the population 
- so what we do is we reduced the value in the sample so that when we reduce the denominator the entire value of the variance and the standard deviation kind of increase a little so it balances out and makes it closer to the population standard deviation and population varaince

## Skewness

- Skewness refres to distortion or asymmetry in a symmetrical bell curve, or normal distribution , in a set of the data 
- if the curve is shifted to the left or to the right , it is said to be skewed 
- skewness can be quantified as a represntation of the extent to which a given distribution varies from a normal distribution 
- A normal distribution has a skew of zero

 - A Normal distribution has no skew so for thus mean , median and mode will be the same value and this is the perfectly symmetrical distribution 
- Then we have negaitively skewed it is skewed towards the negative direction, towards the left direction 
- Then we have Positively  skewed it is skewed towards the positive direction, towards the right hand direction 
- In negative skewed mode will be higher than median and median will be higher than mean value
- while  in Postive  skewed mean will be higher than median and median will be higher than mode value

## Infential Statistics

**Descriptive Statistics**
-It consist of the collection organization summerization and the presentation of the data (it describe the situaion as it is)

**Infrential statistics**
- It consist of making infrences from samples to populations , hypothesis tetsing , determining relationship between variables and making predictions (it is based on probability theory )
- In case of infrential statistics we will be taking up a sample and based on that sample we make infrences regarding the population data because usually we will not be able to find out everything about the population, Finding out the measure regarding the population is very tedious task
- For example we are talking about salaries of all the people in the world , Now calculating the salaries for all the people in the world will take lot of time so instead of calculating and summing up the salaries of all people of the world what we do is we take small sample of people from all the world and based on these samples we try to predict we try to find out salaries which will be belonging to the all the people of the world so this is what infential statistics is
- we will have some sample data and based on the sample data we will try to find out about the population 
- again we will try to compare different samples and find out if they belong to the population or not and if there is some relationship between these two samples  or not  so these kind of things , which when we do those are called infrential statistics

### Why we need Infrential statistics

- Descripive statistics describe the data 

- infrential Statistics allows us to make prediction(infrences) from the data
- With infrential statistics we take data from samples and generalize about the pupulation 
- we can take mean of the samples and then try to find out what would be the mean of the population
- For exaple we have bit of sample of people who are diabetic , so from their diabetes level, i can find out about the diabetes level of people all around the population

- ** Infrential statistics provide a way of going from a "sample" to "population" inferring the parameters of a population from data on the statistics of the smaple
