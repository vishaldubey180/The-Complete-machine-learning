```python
age = np.random.choice([15,20,30,45,12,'10',7,'Mising'],50)
fico = np.random.choice(['100-150','150-200','200-250','250-300'],50)
city = np.random.choice(['mumbai','delhi','chennai','kolkata'],50)
id=np.arange(50)
rating = np.random.choice(['excellent','good','bad','pathetic'],50)
balance = np.random.choice([10000,20000,30000,40000,np.nan,500000,60000],50)
children = np.random.randint(high=5,low=0,size=(50,))
```

```python
data = pd.DataFrame({'ID':id,'age':age,'fico':fico,'city':city,'rating':rating,'balance':balance,'children':children})
```

- fico is also string so we can not perform any operation on that so we have to convert that into numeric value

- **1. First we convert age column into numeric column** 

```python
data['age']= pd.to_numeric(data['age'])
```

- we are getting this error because our column contain "missing" word which is raising another exception
- so at postion 2 it is getting missing hence it is not able to understand what to impute there 
- so to resolve this issue we use error ='coerce' 
- what it does is whenever it find error it will convert that into nan

```python
data['age']= pd.to_numeric(data['age'],errors='coerce')
```

```python
data['age']
```

- we can se that missing values are converted into nan values 
- let's check datatype again

```python
data.dtypes
```

```python
# Now we can see that age is converted into float values
```

```python
data['const_var']=100
data['balance_log']= np.log(data['balance'])
data['age-children_ratio']=data['age']/data['children']
```

```python
data.head(20)
```

- first we created const-var and provide every value as 100
- next we created balance log so what it does it take log values of balance data and create new column
- where balance is nan the balance log is also nan
- so when we impute values in balance column so we have to impute in balance_log column also
- Next we created age-children-ration by dividing age by children

We can do many complex algebric calculation as well to create add/new columns to the data , notice that when a missing value is involved in any calculation , the result is also a missing values .

Let's say we didn't want to this to happen and wanted to impute missing values before proceeding with creating that ratio variable

we can use function isnull to identify values which are using

```python
data.isnull().sum()
```

```python
data['age'].isnull()
```

This gives us a logical array which we can use to isolate missing values and then change them to whatever value we want to impute with

To quickly count howmany missing values are there , we can add sum on top of previous command 

```python
data['age'].isnull().sum()
```

**let's impute these missing values**

```python
data.loc[data['age'].isnull(),'age']#it will give the missing values from age column
```

```python
data.loc[data['age'].isnull(),'age']= data['age'].mean()
```

```python
data['age_children_ratio']=data['age']/data['children']
```

```python
data.isnull().sum()
```

Since age doesen't have missing values anymore , we don't see any missing value in the ratio column, however we do see inf , which is short for infinity which occurs wherever we are dividing by 0
