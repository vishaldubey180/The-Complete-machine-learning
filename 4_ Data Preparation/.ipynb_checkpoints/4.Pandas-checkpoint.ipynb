```python
import numpy as np
import pandas as pd
```

- Now let's say we want to sort by dataframe with respect to column A
- So we can use sort_values

```python
df.sort_values('A',inplace=True)
```

- if we don't provide inplace=true then changes will not save 
- to change save we have to provide inplace=True
- if we want to provide inplace=True then we have to write code like this
- df = df.sort_values('A')

```python
df
```

- For sorting in descending manner we need to use option ascending to false

```python
df.sort_values('A',ascending=False, inplace=True)
```

```python
df
```

- we can sort by multiple column also , sorting by next column in sequence happen within the groups formed as result for sorting by previous columns

```python
df.sort_values(['B','C'],ascending=[True,False])
```

- This sorts the data by column B in ascending order but sorting by column c happen in descending manner , although we haven't taken an explicit example for character data incase of character data , sorting happens in dictionary order

- many times our datasets need to be combined by keys instead of simply stacking them vertically or horizontally

```python
df1 = pd.DataFrame({'cusiid':[1,2,3,4,5],
                  'product':['radio','radio','fridege','fridege','phone']})
```

```python
df2 = pd.DataFrame({'cusiid':[3,4,5,6,7],
                  'stste':['UP','UP','UP','MH','MH']})
```

- nOW LET'S DISCUSS HOW WE CAN MERGE DIFFERENT DATAFRAMES
- WE CAN USE MERGE FOR GETTING DIFFERENT KIND OF JOIN , KEY HERE BETWEEN TWO COLUMNS IS CUSTID

```python
df1
```

```python
df2
```

- here we will merge df1 and df2 on the basis of cutid

```python
pd.merge(df1,df2,on=['cusiid'],how='inner')
```

- Here we are doing inner join 
- so what happen in inner join we take only common rows are between df1 and df2 on the basis of custid

```python
pd.merge(df1,df2,on=['cusiid'],how='outer')
```

- Here we are doing outer join 
- so what happen in outer join it will gives all the rows from the df1 and df2

```python
pd.merge(df1,df2,on=['cusiid'],how='right')
```

- Here we are doing right join 
- so what happen in right join it will take all the rows from right side and common rows between df1 and df2

```python
pd.merge(df1,df2,on=['cusiid'],how='left')
```

- Here we are doing left join 
- so what happen in left join it will take all the rows from left side and common rows between df1 and df2

```python
df3 = pd.DataFrame({'cusiid':[1,2,3,4,5,np.nan],
                  'product':['radio','st','fridege','laptop','phone','other']})
```

# pandas Row-Column Transformaion

```python
tips = sns.load_dataset('tips')
```

A question we might want to ask what is the male to female ratio on different days of the week?

To do this we might start with a groupby

```python
tips_gb = tips.groupby(['day', 'sex']).agg({'size':'sum'})
```

```python
tips_gb
```

So we are getting somewhere , but it is a bit hard to tell the number of male and female visitors by looking at it, you might want to do more columnwise operation comparing the male to female visitors

So what you might want to do is tke the values in the column sex and make them into column. This is where unstacking comes in 

## Unstack

```python
tips_us = tips_gb.unstack()
```

```python
tips_us
```

 When we do group by there is some indexing which makes us difficult to to explain
 
 Unstacking can solve this problem
 
 

- Notice we basically moved an index to the columns 

```python
# we can do the same with the days of the week
tips_us.unstack(0)
```

```python
tips_gb.unstack(1)
```

```python
tips_us.columns
```

## What is Melting

That is about it when it comes to stacking and unstacking . Anything we can do with melting and pivoting can be done with stacking and unstacking . Let's do a single example froom pandas

```python
cheese = pd.DataFrame({'first':['max', 'Marry'],
                      'last':['joe', 'carl'],
                       'height':[5.5, 5.8],
                       'weight':[110, 90]})
```

```python
cheese
```

```python
# melt does stacking in one operation
```

```python
cheese.melt(id_vars=['first','last'])
```

To do this with stacking we just need to do it in two steps

```python
cheese.set_index(['first','last'], inplace=True)
cheese
```

```python
cheese.stack().reset_index()
```

We have used melt before , but after getting a better understanding of stack and unstack we have found them more versitile and stable than the former . so why learn both!!

# Dummy variable 

there is one final way to transform the values in a column into headers and this is called making dummy variables 

making a dummy variable will take all the k distinct values in one column and make k column out of them 

let;s look at an example below

```python
tips.head()
```

```python
pd.get_dummies(tips.head(), columns=['sex'])
```

We can notice that the sex column was split into the sex_male and sex_female column. when the sex is female the sex_female is 1 and 0 otherwise . And similarly for the sex_male column 

This can be very useful for ML models and doing some types of analysis
