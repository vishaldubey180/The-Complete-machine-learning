- here we are getting just a single line here 
- it doesen't ahve any name or label
- Let's provide lable and title to the graph

```python
# x axis values
x=[1,2,3]
# corrsponding y axis values
y=[2,4,1]

# plotting the points
plt.plot(x,y)


#naming the x-axis
plt.xlabel('x-axis')
# naming the y axis
plt.ylabel('y-axis')

# giving a title to our graph
plt.title('our first graph')

# plot grid
plt.grid(True,color='k')

# function to show the plot
plt.show()
```

**Let's plot two line instead of 1**

-----------------------------------

```python
# line 1 points
x1 = [1,2,3]
y1 = [2,4,1]
# plotting the line 1 points
plt.plot(x1,y1, label='line1')

# line 2 points
x2 = [1,2,3]
y2 = [4,1,3]
# plotting the line 2 points
plt.plot(x2,y2, label='line2')


# naming the x-axis
plt.xlabel('x-axis')
# naming the y-axis
plt.ylabel('y-axis')
# giving title to the graph
plt.title('two lines on same graph')

# show legend on the plot
plt.legend()

# function to show the plot
plt.show()
```
