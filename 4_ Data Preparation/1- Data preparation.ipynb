# Problems With Variables 

**1.Irrelavant column**

- When we rae working with different variable and different features we need to make sure that the feature which we have are representing good amount of data and also they should not have any irrelevant information
- for ex let's say we have some irrelevant column like if we are talking about loan approval application now application number is completely irrelevant column because the application number column will not make any difference in approval or rejection of the application 

**Catagorical column**

- Now let's say we have catagorical column which have lot of categories, so in this particular situation when there are lot of cataegories we can not really convert it into one-hot-encoding 
- so in this particular case we can select either top 7 or top 10 categories or we can divide the categories into subgroups 
- So let's say we have categories city name or state so what we can do if we have 50 cities then we can may be choose most influential cities to 10 or 5.
- or we can group the cities as matroplolitan cities nd so on so that we have only two or three cities in which multiple city can lying 


**Outliers - ANomly detection**

- There could be certain outliers or some kind of data which is not really important in nature
- So what we can do is we can use anomly detection or we can use outliers detection using boxplot and get rid of outliers 
- Instead of outliers we can impute certain values 
- we can replace these values with the mean of the data or the median of the data or the mode of the data in case it is categorical data 


**Missing Value**

- In case there are any missing values in that particular case we can add some other data 
- So in case of missing value if there are few values which are missing then we can take a mean or median of the data and we can impute those values in the place of missing values 
- In case the missing values are huge in number let's say 90% of the data is missing in that case we can directly get rid of the column itself because imputing values will not be fruitful for us because we would be creating false value in that situation 

**Noise**

- When there is some noise that is modification of the original values these noise can look like original input data but has some kind of falut in it and it is very hard to detect 
- so to avoid noise what we can do is we can make sure the data from our source is being pulled correctly that is the only solution which we can have for noise
