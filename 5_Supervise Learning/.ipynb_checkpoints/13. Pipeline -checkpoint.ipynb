So basically pipeline will hepl us to automate the entire machine learning workflow 

so next timw when we get new data we put that into pipeline and it will do all the task and will give us output

So in pipeline we have training data which goes into pipeline 

first it will be cleaned , then it will be normalize and then it will impute the missing value then feature enginerrring will be done after that we can apply any of the model which we want
