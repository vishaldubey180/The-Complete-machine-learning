# Ensemble learning

- It Combine multiple weak algorithm to form strong model 
- Using ensemble method allows to produce better predictions as compared to single model 

So instead of creating a large decision tree which is very prone to being overfitting we can create small decision trees and combine them together in such way that we will have strong ML model which will not overfit

# Ensembel Learning method 

- Bagging is first ensembel learning method 

- SO what bagging does is that bagging will create small models here model1,model2,model3.model4,model5. These are small models which are created randomly 

- So we train each model individually and then we will take their votes and from their votes we will decide the majority vote will be the answer or the average of the values will be the answer of the problem 

- SO the predicting values will be the average or the majority votes from the above five models because these models are completely unrelated with each other so we can gereate these models parallely and then simply combine them by taking a majority of them 

2. Boosting

- In boosting we craetes sequential models
- First we will create model1, then model2, then model 3 ,model4 and model5

- Now all models will get x value as input 

- the first model will get input value  x and y and it will try to predict y value but model being the weak model it will not be able to identify all the aptterns so it will predict something which is not close to y 
- So the prediction which is amde by model 1 let it be y1
- Now the error which we are having from model 1 will be y-y1

- Now for second model again we get the input as x but this time instead of predicting y we will be predicting error value 

- So now we will try to imrove the model by simply predicting the difference between the actual value and the value which we are getting from model 1 so we will try to imrove that 

- So now we will get another prediction and that prediction will somehow imrove the value more so now the error is y-(y1+y2)

- So now there is lesser amount of error left 

- Similarly we will do with model 3 that is we try to predict error genrated by model2 and model 3 try to reduce those error by learning more or observing more pattern and imrove the prediction 

- we will do same thing with model 4 and model 5 

- And finally we will have the value y-(y1+y2+y3+y4+...) until this value become 0 

- now entire term will be become zero because we are imroving the model one after other 

- Incase of bagging the models are created paralally 
- in boosting the models are created sequentially 
- In boosting each model is trying to improve previous model 
- in bagging the models are independent and does not have any relationship between them 
- in bagging we take a average of majority votes 
- in boosting we try to reduce error of the previous model 
- bagging tries to reduce variance of smaller model 
- boosting will tries to reduce the bias of the models
