# Unsupervise learning

- In case of unsupervise learning, no target value is provided

- In case of unsupervise learning , the focus is on grouping dissimilar entities by finding structure 
- here we are not concerned with finding out something and predicting something for the future , we just want to group the data with similar entities together
- In case of unsuperwise learning we have to find Anomlies that we can used to find something which is out of packet, something which is unusual 
- Unsupervise learning helps in dimensiality reduction
