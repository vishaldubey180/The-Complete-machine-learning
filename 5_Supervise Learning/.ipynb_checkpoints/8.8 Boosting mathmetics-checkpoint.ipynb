# Boosting mathametics

 Maths behind the boosting algorithm basically consist of 3 steps 
 
 First step
 
 - An intial model F0 which is trying to predict the target variable y. This mdoel will be associated with residual or error  (y-F0)
 
 - Now a new model h1 will be generated now this model will be fit to residual from the previous step so now the target value will be actually y-F0
 
 - Now F0 and h1 are combined to give f1, the boosted version of F0. The mean squared error from f1 will be lower than that from F0
 
 - If we will look in below digram then first model will be f0 and all the other next model will be h1,h2,h3 i.e intermediate model and the combination of these model so the first model + second model i.e f0+h1 =f1 model AND F0+H1+H2 WILL Be the f2 model and f0+h1+h2+h3 will be f3 model 
 
 - Now F0 and h1 are combined to give f1, the boosted version of F0. The mean squared error from f1 will be lower than that from F0. Why because f0 has one decision stumps ad h1 has one decision stumps f0 would have given some residual and h1 would have improvised those error so when we comnine them the resultant residuals the resultant error will be lower in comparison to the previous error if it is not lower then we will not consider that particular stump 

Now let's consider the prediction at iteration t will be ft so this will be equal to function of x 

So prediction at any particular iteration is Ft which is final combined model i.e at any iteration f0 and h1 is known as Ft while the small weak learner  are each called ft and these will be added from 0-number of iteration  

- So ft will be weak learner which we are adding and Ft will be final strong learner 

- So the loss value or the error value is calculated by y- predicted value (Ft(X))\

- So here we have calculated loss value as J or cost function so cost function will be entire loss 
- So loss function is at any point what is the error and error will be y-Ft
- So when we subtract predicted value fromthe actual value it is called as error and when we combine them it is called as cost 


- Now we are trying to find out the change in cost w.r.t the model , at every step how much the cost is changing 

- So it will be delJ w.r.t to Ft(X)

- Because here the only thing which is changing is the function F. x and Y are already constant so only change is Ft hence we diffrentiate delj w.r.t to Ft

Now the change which should be brought to a particular model is del j w.r.t ft

Now the next model will be ft+1 and it will be sum of current model + the change which w e want to bring to that model 

n - learning rate 

means how fast we want to change our  model 

and final model will be the previous model with all the layers + new weak stump which has been added to it
