# Boosting and Adaboost

- As we know in boosting we have sequential model where each of the first model will be learning from the input data .i.e it will learn from x and try to predict y 
- the error generated by first model will be pushed into the next model to be predicted 
- so the target for the second model will actually be the error from the first model 
- so it will keep on improvising the previous model in the next iteration 
- So the last model we would expect that the last model has no error present in it 
- now if we will have no error present in the last model , then tha means we would have overfiteed the data somewhere 
- so we should know where we need to stop and we should also know that there has to be some kind of regularization which needs to be applied on this boosting algorithm 
- so the major points to be remember about boosting is that these model are generated in a sequential order in comparison to bagging where in bagging the models are created parallely 

- And boosting is craeted by diffrent decision tree stump again 

- Now these decision trees stumps does not try to fit data to any specific patern like linear model which would always try to fit data to linear equations 
- So here there is no specific pattern which we are trying to fix we are jut trying to create small models that each of these small models would be explored(expert) on something we just want to create more and models which will be able to emphasize on one specific  pattern

- so each model will learn something and push it forward to the other model , and another model will try to improvise on the pattren which were missed out by the previous model .

- So there will be certain amount of weightage which will be given to the model one the items in the model1 which model 1 could not predict  

- So the values which were misclassified or not predicted properly by model one will be improvised in model 2 by giving higher weightage  to then 

- So this is how each model will try to learn some part of the pattern which are present 

- And somehow entire combination will be able to learn all the pattren this way and will be able to predict better in comparison to a normal decision tree 

- So here the input varibles remain the same in all weak learners 

- While the target variable is changed to the error value from the previous model 

- The target value for the first model is Y

- And target value for the second model will be y-wl1,wl2 etc
