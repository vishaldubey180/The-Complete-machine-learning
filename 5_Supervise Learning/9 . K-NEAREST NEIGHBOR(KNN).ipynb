# Properties of KNN

1. KNN can be used for both classification as well as  regression prdictive problem 

     In case of classification we will take the nearest k and take majority vote of the points and then decide which class the points should belong 
     When we are working with regression we will consider those points and will take weighted average of the avlues 
     a weighted average will allow to regularise the value because some value might be close and some points might be far up  
     
      so aweighted average on the basis of distance will allow to modulate the value in case of finding out the regression value 
      
2. It is mainly used for classification predictive problem 

   it is distance based algorithm and it might ignore some pattern because it is looking for nearest point so it is looking in a circular region around it 
   
3. It is called a lazy learning algorithm 

    Because there is no specialized training phase 
    
    it uses all the data for training while classification 
    (so it will not need explicit training time period when we are actually making a prediction we can actually provide all the data points and these data points will itself act as a training model ,
    if we remove few datapoints then it will be classified differently and if we add another datapoint then it will classify differently )
    That is why it is called as lay learning 
    
4. It is non-parametric learning algorithm 

     It doesn't assume anything about underlaying data 
     and it find results based on closely matching data points 
     distance based algo hence scaling is required
