```python
##print board:
def printbrd(brd):
    print('''\t |     |     
       {} |  {}  | {} 
         |     |     
    _________________
         |     |     
       {} |  {}  | {} 
         |     |     
    _________________
         |     |     
       {} |  {}  | {} 
         |     |     '''.format(brd[0],brd[1],brd[2],brd[3],brd[4],brd[5],brd[6],brd[7],brd[8]))
```

```python
def wincheck(brd,marker):
    boardstate = {index for index in range(9) if brd[index]==marker}
    #print(boardstate)
    for case in winlist:
        if boardstate.issuperset(case):
            return 'win'
```

```python
#Main loop for Game:
winlist=[{0,1,2},{3,4,5},{6,7,8},{0,3,6},{1,4,7},{2,5,8},{0,4,8},{2,4,6}]
def gameloop():
    brd=['#']*9
    nums = ['1','2','3','4','5','6','7','8','9']
    winlist=[{0,1,2},{3,4,5},{6,7,8},{0,3,6},{1,4,7},{2,5,8},{0,4,8},{2,4,6}]
    gameon = True
    player = True
    while gameon:
        if player: #check for Player 1   
            val=input('Player 1: Please enter 1-9 to place 0, type anything else to end game')
            marker='0'
            if val in nums:
                print('entered number = '+ val)
                nums.remove(val) 
                brd[int(val)-1]=marker
                printbrd(brd)
                if wincheck(brd,marker)=='win':
                    print('Player 1 is the winner!!!')
                    gameon = False
            else:
                print('Do you want to Quit?')
                quitter=input('Y/N').lower()
                if quitter == 'y':
                    gameon=False
                else:
                    continue
                          
        else:    #check for Player 2
            val=input('Player 2: Please enter 1-9 to place X, type anything else to end game')
            marker='X'
            if val in nums:
                print('entered number = '+ val)
                nums.remove(val)
                brd[int(val)-1]=marker
                printbrd(brd)
                if wincheck(brd,marker)=='win':
                    print('Player 1 is the winner!!!')
                    gameon = False
            else:
                print('Do you want to Quit?')
                quitter=input('Y/N').lower()
                if quitter == 'y':
                    gameon=False
                else:
                    continue
        player = not player
```
